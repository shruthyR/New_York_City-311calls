## Whether the average response time across complaint types is similar or not (overall)

```python
from scipy.stats import f_oneway
```

```python
data=nyc1['Complaint Type']
time=nyc1['Request_Closing_Time']
```

```python
location1=data=='Noise - Street/Sidewalk'
location2=data=='Blocked Driveway'
location3=data=='Illegal Parking'
```

```python
new_location1=time[location1].dropna()
new_location2=time[location2].dropna()
new_location3=time[location3].dropna()
```

```python
print(f_oneway(new_location1,new_location2,new_location3))
```

### p-value is greater than 0.05, so it significes we reject the null hypothesis. the average response time across complaint type is not similar

```python
frequency=pd.crosstab(nyc1['Complaint Type'],nyc1['Location Type'])
frequency
```

## Are the type of complaint or service requested and location related?

```python
from scipy.stats import chi2_contingency
```

```python
stat, p, dof, expected=chi2_contingency(frequency)
print(p)
```

### p-value is not greater than 0.05, so we accept null hypothesis. And the average response time across complaint type is not similar
